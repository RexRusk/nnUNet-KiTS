```python
def maybe_mkdir_p(directory):
    directory = os.path.abspath(directory)
    splits = directory.split("/")[1:]
    for i in range(0, len(splits)):
        if not os.path.isdir(os.path.join("/", *splits[:i + 1])):
            try:
                os.mkdir(os.path.join("/", *splits[:i + 1]))
            except FileExistsError:
                # this can sometimes happen when two jobs try to create the same directory at the same time,
                # especially on network drives.
                print("WARNING: Folder %s already existed and does not need to be created" % directory)
```

```python
import shutil
# Source directory containing the .json files
source_dir = "/content/nnUNet/DATASET/nnUNet_preprocessed/Dataset040_KiTS/nnUNetPlans_3d_lowres"

# Destination directory where you want to copy the files
destination_dir = "/content/nnUNet/DATASET/nnUNet_preprocessed/Dataset040_KiTS/"

# List of .json files to copy
files_to_copy = ['dataset.json', 'dataset_fingerprint.json', 'nnUNetPlans.json']

# Loop through the list of files and copy them one by one
for file in files_to_copy:
    source_path = os.path.join(source_dir, file)
    destination_path = os.path.join(destination_dir, file)
    shutil.copyfile(source_path, destination_path)
    os.remove(source_path)
```
